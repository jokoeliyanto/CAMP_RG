#### 1.1.3 Modus

```python
dict_lang_count = {}

for value in lang:
    if value not in dict_lang_count.keys():
        dict_lang_count[value] = 1
    else:
        dict_lang_count[value] += 1

print("Frekuensi masing-masing nilai:", dict_lang_count)
print("Nilai dengan jumlah terbanyak:",
      max(dict_lang_count.items(), key=lambda items: items[-1]))
```

```python
dict_lang_count = {}

for value in lang:
    if value not in dict_lang_count.keys():
        dict_lang_count[value] = 1
    else:
        dict_lang_count[value] += 1

max_level_freq = max(dict_lang_count.values())
modes = {
    level: mode
    for level, mode in dict_lang_count.items()
    if mode == max_level_freq
}

print("Frekuensi masing-masing nilai:", dict_lang_count)
print("Modus:", modes)
```

### 2.4 Visualisasi

#### 2.4.1 Visualisasi Kolom Kategori Umur

```python
age_category_order = df.AgeCategory.value_counts().index
```

```python
df.AgeCategory.value_counts()
```

```python
age_category_order
```

```python
age_category_order = df.AgeCategory.value_counts().index

plt.figure(figsize=(12, 6))
sns.countplot(
    x="AgeCategory",
    data=df,
    order=age_category_order,
    palette=["#29B5BF"] + (len(age_category_order)-1)*["#777777"]
)
plt.grid(False)
plt.show()
```

Dari visualisasi di atas, terlihat bahwa kategori umum **65-69** paling banyak dijumpai.

Untuk kemudahan visualisasi fitur yang lain, kita akan menggunakan fungsi berikut.

```python
def plot_count(x, data=df, palette=None, figsize=(12, 6)):
    column_order = data[x].value_counts().index

    plt.figure(figsize=figsize)
    sns.countplot(
        x=x, data=data, order=column_order,
        palette=palette
    )
    plt.grid(False)
    plt.show()
```

```python
plot_count('PhysicalActivity', data=df, palette=None, figsize=(12, 6))
```

#### 2.4.2 Visualisasi Kolom Race

```python
race_order = df["Race"].value_counts().index
plot_count("Race", palette=["#29B5BF"] + (len(race_order)-1)*["#777777"])
```

Terlihat bahwa dari 6 suku yang ada, `White` menjadi yang paling banyak. Sedangkan untuk `Asian` berada pada urutan 2 terakhir.

#### 2.4.3 Visualisasi Kolom General Health

```python
gen_health_order = df["GenHealth"].value_counts().index
plot_count(
    "GenHealth",
    palette=["#29B5BF"] + (len(gen_health_order)-1)*["#777777"]
)
```

Dari visualisasi di atas, bisa disimpulkan bahwa kebanyakan responden sangat percaya bahwa kondisi kesehatan mereka secara umum sangat baik.

#### 2.4.4 Visualisasi Kolom Jenis Kelamin

```python
sex_order = df["Sex"].value_counts().index
plot_count(
    "Sex",
    palette=["#29B5BF", "#FFE653"]
)
```
